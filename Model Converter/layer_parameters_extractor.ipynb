```python
import tensorflow as tf

def print_model_parameters(model):
    for layer in model.layers:
        print("Layer Name:", layer.name)
        print("Layer Trainable:", layer.trainable)
        
        # Access layer weights
        layer_weights = layer.get_weights()
        if len(layer_weights) > 0:
            for i, weight in enumerate(layer_weights):
                print("Weight", i, "Shape:", weight.shape)
                print("Weight", i, "Values:\n", weight)
        
        # Access layer biases
        layer_biases = layer.get_weights()
        if len(layer_biases) > 1:
            for i, bias in enumerate(layer_biases):
                print("Bias", i, "Shape:", bias.shape)
                print("Bias", i, "Values:\n", bias)
        
        # Additional parameters specific to certain layer types
        if isinstance(layer, tf.keras.layers.Conv2D):
            kernels = layer.get_weights()[0]
            for i in range(kernels.shape[-1]):
                print("Kernel", i, "Shape:", kernels[..., i].shape)
                print("Kernel", i, "Values:\n", kernels[..., i])
        elif isinstance(layer, tf.keras.layers.Dense):
            weights = layer.get_weights()[0]
            biases = layer.get_weights()[1]
            print("Weights Shape:", weights.shape)
            print("Weights Values:\n", weights)
            print("Biases Shape:", biases.shape)
            print("Biases Values:\n", biases)
        
        print()  # Empty line for separation

# Load or build your TensorFlow model
model = tf.keras.models.load_model('D:\Programing\Projects\ESP32-TFLite-Knock-Detector\Train\pruned_model.h5')

# Call the function to print model parameters
print_model_parameters(model)
```
